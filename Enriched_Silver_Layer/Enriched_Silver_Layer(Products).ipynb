```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("parquet").load("abfss://bronze@azuredatabricksjesslake.dfs.core.windows.net/products")
df.display()
```

```python
df = df.drop("_rescued_data")
```

```python
df.display()
```

```python
df.createOrReplaceTempView("products")
```

```python
df =df.withColumn("discounted_price", col("price") * 0.9)
df.display()
```

```python
df.write.format("delta").mode("append").option("path", "abfss://silver@azuredatabricksjesslake.dfs.core.windows.net/products").saveAsTable("dtbproject.silver.products")
```
