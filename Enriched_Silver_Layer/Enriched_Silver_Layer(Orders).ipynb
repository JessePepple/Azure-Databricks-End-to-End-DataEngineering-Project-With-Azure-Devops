# Data Reading

```python
df = spark.read.format("parquet").load("abfss://bronze@azuredatabricksjesslake.dfs.core.windows.net/orders")
```

```python
df.display()
```

```python
df = df.drop("_rescued_data")

df.display()
```

```python
from pyspark.sql.functions import *
```

```python
df = df.withColumn("order_date", to_timestamp(col("order_date")))
df.display()
```

```python
df = df.withColumn("year", year(col("order_date")))
df.display()
```

Since this is our Fact Table not much transformation is needed

```python
df.write.format("delta").mode("overwrite").option("path", "abfss://silver@azuredatabricksjesslake.dfs.core.windows.net/orders").saveAsTable("dtbproject.silver.orders")
```
