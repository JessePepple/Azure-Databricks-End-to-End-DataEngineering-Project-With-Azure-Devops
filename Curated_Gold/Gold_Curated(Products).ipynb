DLT Pipeline

```python
import dlt
from pyspark.sql.functions import *
```

## Set Expectations

```python
expectations = {
    "rule 1" : "product_id IS NOT NULL",
    "rule 2" : "product_NAME IS NOT NULL"
}
```

Our Table

```python
@dlt.table()

@dlt.expect_all_or_drop(expectations)
def DimProducts_stage(): 

  df = spark.readStream.option("skipChangeCommits","true").table("dtbproject.silver.products")

  return df
```

## Streaming View

```python
@dlt.view 

def DimProducts_view():

    df = spark.readStream.table("DimProducts_stage")
    return df 
```

```python
dlt.create_streaming_table("DimProducts")
```

```python
dlt.apply_changes(
  target = "DimProducts",
  source = "DimProducts_view",
  keys = ["product_id"],
  sequence_by = "product_id",
  stored_as_scd_type = 2
)
```
